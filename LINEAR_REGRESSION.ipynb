```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 2, 2.5, 4, 5])
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x, y, color='black')
plt.plot(x, y_pred, color='blue')
plt.title("linear regression")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
print("inter (Î²Î¸):", model.intercept_)
print("coefficient (Î²1):", model.coef_)
```

```python
X = np.random.rand(100, 1) * 100
Y = 2 * X + np.random.randn(100, 1) * 10
```

```python
plt.scatter(X, Y, color='blue')
plt.title("Generated Dataset of 100 Numbers")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='blue')  # Actual data points
plt.plot(X, Y_pred, color='red')  # Regression line
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
print("Intercept (Î²0):", model.intercept_)
print("Coefficient (Î²1):", model.coef_)
```
